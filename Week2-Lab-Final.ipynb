```python
#activity B

number_of_runs=0 #defining variable and assigning value of 0
total_time=0 #defining variable and assigning value of 0

while True: #while loop to run untill true/ break
    #taking input in variable for run time or 'q' for quit
    one_run=input("Enter 10 km run time , or 'q' to exit: ") 
    # if one_run=q then break loop
    if one_run=='q':
        break
    #if one_run is not equal to q execute below block
    else:
        #incrementing number of runs by 1 every time loop runs
        number_of_runs += 1
        #adding all entered run times in variable one_run into variable total_time
        total_time += float(one_run)

#calculating average into variable average_time
average_time=total_time/ number_of_runs
#print command printing with format atribute where text can be written with variables inside {} brackets
print(f" Average of {average_time}, over {number_of_runs} runs")
```

```python
#linera algebra
import numpy as np
x=np.array([[1.,2.,3.],[4.,5.,6]])
print("==== X array =====")
print(x)
y=np.array([[6.,23],[-1,7],[8,9]])
print("==== Y array=====")
print(y)
print("====x.dot(y)=====")
print(x.dot(y))
print("====np.dot(x,y)=====")
print(np.dot(x,y))
print("====np.dot(x,np.ones(3))=====")
print(np.dot(x,np.ones(3)))
print("====using @ as infix operator=====")
print(x @ np.ones(3))
```
